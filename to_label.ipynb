## 데이터에 라벨을 붙이는 작업

직접 만든 데이터에 라벨을 부여한다. 


### 1. txt fileo to dataset
txt 파일 여러개를 모두 합치고 라벨까지 부여한다. 

```python
import os
```

```python
original_file_path = [ "dataset/tfdata/" + file_name for file_name in os.listdir("dataset/tfdata/")]

after_file_path='dataset'
file_name='my_data_0702.csv'
sentences = []
labels = []
print(original_file_path)
```

```python
# 번호. 문장의 형태로 변경
def arrage_file(original_paths) :
    for cur_file_name in original_paths :
        try :
            with open(cur_file_name.replace('labeldata/0702new', 'tfdata', 1), 'w', encoding='utf-8') as wf :
                with open(os.path.join(cur_file_name), 'r', encoding='utf-8') as f :
                    cnt = 0
                    for line in f :
                        num, _, sentence = line.partition(". ")
                        if num.isdigit() : 
                            wf.write(str(cnt) + ". " + sentence)
                            cnt += 1
                            
        except Exception as e:
            print(f"{cur_file_name} has error {e}")
```

```python
arrage_file(original_file_path)
```

```python

# 데이터에 라벨을 붙여서 하나의 파일로 만든다.
def set_label(original_paths) :
    for cur_file_name in original_paths : 
        try :
            with open(os.path.join(cur_file_name), 'r', encoding='utf-8') as f :
                for i, line in enumerate(f, start=1) :
                    if i == 1 : 
                        first_line = line
                        label,_, title = first_line.partition(". ")
                        label = int(label)
                    elif i == 2 : 
                        continue
                    else : 
                        line_num, _, sentence = line.partition(". ")
                        if sentence == '' : continue
                        sentences.append(sentence.rstrip('\n'))
                        labels.append(label)
        except Exception as e:
            print(f"{cur_file_name} has error {e}")
```

```python
set_label(original_file_path)
print(sentences)
print(labels)
```

```python
# sentences와 labels를 합친 csv 데이터 만들기
import csv
def save_to_csv(sentences, labels, file_path) :
    field_names = ['id', 'sentence', 'label']

    with open(file_path, 'w', newline='', encoding='utf-8') as csvfile : 
        writer = csv.DictWriter(csvfile, fieldnames=field_names, quoting=csv.QUOTE_MINIMAL)

        writer.writeheader()

        for id, (sentence, label) in enumerate(zip(sentences, labels)) :
            writer.writerow({'id' : id, 'sentence' : sentence, 'label' : label})

save_to_csv(sentences, labels, after_file_path + '/' + file_name)
print(f"Data has been written")
```
